```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st


d_small=np.loadtxt("emission_small.txt")
d_large=np.loadtxt("emission_large.txt")

std_emi=1

plt.hist(d_small,histtype="step",bins=30)
plt.hist(d_large,histtype="step",bins=30);
```

```python
xs=np.linspace(0.5,1.4,200)

mean, std=d_small.mean(), d_small.std()/np.sqrt(d_small.size)

print("mean is ",mean)
print("std deviation is ",std)

d_norm=st.t.pdf(xs,10,mean,std)

plt.plot(xs,d_norm)
plt.axvline(1,color="red");
```

```python

mean, std=d_large.mean(), d_large.std()/np.sqrt(d_large.size)
d_norm=st.norm.pdf(xs,mean,std)
d_normstd=st.norm.cdf(std_emi,mean,std)
print(st.norm.cdf(std_emi,mean,std))

plt.plot(xs,d_norm)
plt.axvline(std_emi,color="red")
plt.fill_between(xs,d_norm,where=xs>std_emi,color="green");
```
